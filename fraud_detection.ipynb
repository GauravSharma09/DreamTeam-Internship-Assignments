```python
def classify_time(hour):
    if 5 <= hour < 12:
        return "Morning"
    elif 12 <= hour < 17:
        return "Afternoon"
    elif 17 <= hour < 21:
        return "Evening"
    else:
        return "Night"

df['time_period'] = df['hour'].apply(classify_time)
df.head()
```

```python
sns.countplot(data=df, x='time_period', hue='is_fraud')
```
